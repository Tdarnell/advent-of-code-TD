```python
import math


def extract_digits(num: int) -> list[int]:
    if num == 0:
        return [0]
    # Source - https://stackoverflow.com/a
    # Posted by inspectorG4dget, modified by community. See post 'Timeline' for change history
    # Retrieved 2025-12-05, License - CC BY-SA 4.0
    return [(num//(10**i))%10 for i in range(math.ceil(math.log(num, 10)), -1, -1)][bool(math.log(num,10)%1):]

def find_joltage(digits: list[int]):
    if not digits:
        return 0
    if not isinstance(digits, list):
        raise ValueError("Input to find_joltage must be a list of integers.")
    if not all(isinstance(d, int) for d in digits):
        raise ValueError("All elements in the input list must be integers.")
    max1 = max(digits[:-1])
    max_idx = digits.index(max1)
    max2 = max(digits[max_idx + 1 :])
    max2_idx = digits[max_idx + 1 :].index(max2) + (max_idx + 1)
    if max_idx > max2_idx:
        return max2 * 10 + max1
    return max1 * 10 + max2


inputs = ["987654321111111", "811111111111119", "234234234234278", "818181911112111"]
for i in inputs:
    input_digits = [int(_) for _ in i]
    print(find_joltage(input_digits))
```

```python
def find_joltage(digits: list[int], num_maxs=2):
    if not digits:
        return 0
    if not isinstance(digits, list):
        raise ValueError("Input to find_joltage must be a list of integers.")
    if not all(isinstance(d, int) for d in digits):
        raise ValueError("All elements in the input list must be integers.")
    output = 0
    current_slice_start = 0
    for m in range(num_maxs):
        current_slice_end = len(digits) - (num_maxs - (m + 1))
        slice_digits = digits[current_slice_start:current_slice_end]
        next_digit = max(slice_digits)
        current_slice_start = slice_digits.index(next_digit) + current_slice_start + 1
        output = output * 10 + next_digit
    return output


inputs = ["987654321111111", "811111111111119", "234234234234278", "818181911112111"]
for i in inputs:
    input_digits = [int(_) for _ in i]
    print(find_joltage(input_digits, 12))
```
